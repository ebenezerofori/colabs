#Line Charts

```
import numpy as np 
import matplotlib.pyplot as plt
```

```
x = np.linspace(0, 20, 1000) 
```

```
y = np.sin(x) + 0.2 * x 
```

```
plt.plot(x, y) 
plt.xlabel('input')
plt.ylabel('output')
plt.title('my first plot'); # incase it was last command to avoid it being printed
#plt.show()
plt.grid()
```

# Scatter Plot

```
x = np.random.randn(100, 2)
```

```
plt.scatter(x[:, 0], x[:, 1]);
```

```
X = np.random.randn(200, 2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:,0], X[:,1], c = Y);
```

# Histogram 

```
h = np.random.randn(10000)
```

```
#plt.hist(h);
plt.grid();
plt.hist(h, bins=50); # more bars
```
